```python
links
```

```python
documents = []
for i in links:
    r = requests.get(i)
    soup = BeautifulSoup(r.content, 'html.parser')
    sen = []
    for i in soup.find('div', {'class':'read__content'}).find_all('p'):
        sen.append(i.text)
    documents.append(' '.join(sen))
```

```python
documents_clean = []
for d in documents:
    document_test = re.sub(r'[^\x00-\x7F]+', ' ', d)
    # menghapus imbuhan
    document_test = re.sub(r'@\w+', '', document_test)
    # mengubah dokumen ke huruf kecil
    document_test = document_test.lower()
    # menghapus tanda baca
    document_test = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', document_test)
    # mengubah angka
    document_test = re.sub(r'[0-9]', '', document_test)
    # menghapus spasi ganda
    document_test = re.sub(r'\s{2,}', ' ', document_test)
    documents_clean.append(document_test)
```

```python
#menampilkan dokumen setelah prepocessing
documents_clean
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents_clean)
X = X.T.toarray()
df = pd.DataFrame(X, index=vectorizer.get_feature_names())
```

```python
df
```

```python
q1 = input('Inputkan query yang diinginkan : ') #melakukan inputan query
q1 = [q1]
q_vec = vectorizer.transform(q1).toarray().reshape(df.shape[0],)
sim = {}
for i in range(84):
  sim[i] = np.dot(df.loc[:, i].values, q_vec) / np.linalg.norm(df.loc[:, i]) * np.linalg.norm(q_vec)
sim_sorted = sorted(sim.items(), key=lambda x: x[1], reverse=True)

#menampilkan nilai similaritas/kesamaan
for k, v in sim_sorted:
  if v != 0.00:
    print("Nilai Similaritas:", v)
    print(documents_clean[k])
    print()
```

```python
tp = {}
x = 0
for i in range(len(sim)):
  if sim[i] != 0.0:
    x += 1

correct_answer_returned = x
total_answer = 84
total_correct_answer = x

precision = (correct_answer_returned/total_answer)
recall = (correct_answer_returned/total_correct_answer)
f1_score = (2*precision*recall)/(precision+recall)

print(f"precision: {precision:.1f} %")
print("recall   :", recall,"%")
print(f"f1-score : {f1_score:.1f} %")
```
